Понятия "множество" и "последовательность" связаны отношением **общего** и **частного**. 

* **Множество** - это более **общее** понятие, выступающее в роли **родителя**. Оно представляет собой **целое**, совокупность каких-либо объектов, объединенных общим свойством. 

* **Последовательность** - это **частный** случай множества, его **дочерний субъект**. Она представляет собой **часть** множества, в которой важен **порядок** элементов. 

Другими словами, любая последовательность является множеством, но не каждое множество является последовательностью. 

**Пример:**

* **Множество:** Все буквы русского алфавита.
* **Последовательность:** Буквы в слове "алфавит".

В этом примере буквы слова "алфавит" образуют **часть** множества всех букв русского алфавита. При этом, **порядок** букв в слове важен, что делает его **последовательностью**.

Давайте рассмотрим каждое высказывание, построим его отрицание и установим истинность:

1) ∀y ∈ [0; 1] : sgn(y) = 1
   Отрицание: ∃y ∈ [0; 1] : sgn(y) ≠ 1
   Истинность: Высказывание ложно, отрицание истинно. Для y = 0, sgn(0) = 0.

2) ∀n ∈ ℕ > 2 : ∃x, y, z ∈ ℕ : xⁿ = yⁿ + zⁿ
   Отрицание: ∃n ∈ ℕ > 2 : ∀x, y, z ∈ ℕ : xⁿ ≠ yⁿ + zⁿ
   Истинность: Высказывание ложно, отрицание истинно. Это теорема Ферма.

3) ∀x ∈ ℝ ∃X ∈ ℝ : X > x
   Отрицание: ∃x ∈ ℝ ∀X ∈ ℝ : X ≤ x
   Истинность: Высказывание истинно, отрицание ложно.

4) ∀x ∈ ℂ ∃y ∈ ℂ : x > y || x < y
   Отрицание: ∃x ∈ ℂ ∀y ∈ ℂ : x ≤ y && x ≥ y
   Истинность: Высказывание ложно, отрицание истинно. В ℂ нет полного порядка.

5) ∀y ∈ [0; π/2] ∃ε > 0 : sin y < sin(y + ε)
   Отрицание: ∃y ∈ [0; π/2] ∀ε > 0 : sin y ≥ sin(y + ε)
   Истинность: Высказывание истинно, отрицание ложно.

6) ∀y ∈ [0; π) ∃ε > 0 : cos y > cos(y + ε)
   Отрицание: ∃y ∈ [0; π) ∀ε > 0 : cos y ≤ cos(y + ε)
   Истинность: Высказывание истинно, отрицание ложно.

7) ∃x : x ∉ {ℕ, ℤ, ℚ, ℝ, ℂ}
   Отрицание: ∀x : x ∈ {ℕ, ℤ, ℚ, ℝ, ℂ}
   Истинность: Высказывание ложно, отрицание истинно. Все числа принадлежат к одному из этих множеств.

## Множества

```python
#1
# Исходные множества
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
c = {1, 5, 7}

# Объединение
union_ab = a | b   # или a.union(b)
union_ac = a | c
union_bc = b | c

# Пересечение
intersection_ab = a & b  # или a.intersection(b)
intersection_ac = a & c
intersection_bc = b & c

# Разность
diff_ab = a - b   # или a.difference(b)
diff_ac = a - c
diff_ba = b - a
diff_bc = b - c
diff_ca = c - a
diff_cb = c - b

# Симметричная разность
symdiff_ab = a ^ b   # или a.symmetric_difference(b)
symdiff_ac = a ^ c
symdiff_bc = b ^ c

# Вывод результатов
print("Объединение:")
print(f"  a | b = {union_ab}")
print(f"  a | c = {union_ac}")
print(f"  b | c = {union_bc}")

print("\nПересечение:")
print(f"  a & b = {intersection_ab}")
print(f"  a & c = {intersection_ac}")
print(f"  b & c = {intersection_bc}")

print("\nРазность:")
print(f"  a - b = {diff_ab}")
print(f"  a - c = {diff_ac}")
print(f"  b - a = {diff_ba}")
print(f"  b - c = {diff_bc}")
print(f"  c - a = {diff_ca}")
print(f"  c - b = {diff_cb}")

print("\nСимметричная разность:")
print(f"  a ^ b = {symdiff_ab}")
print(f"  a ^ c = {symdiff_ac}")
print(f"  b ^ c = {symdiff_bc}")

print("\nВхождение:")
print(a.issubset(b))  
print(b.issuperset(a)) 

# Дополнительные операции
print("\nДлина множеств:")
print(len(a))
print(len(b))
print(len(c))

print("\nПроверка на пустоту:")
print(a.isdisjoint(b))  
```

Рассмотрим каждую последовательность по очереди:

1) {an}∞n=1 = 2^n - n

a. Монотонность: 
Рассмотрим разность an+1 - an = (2^(n+1) - (n+1)) - (2^n - n) = 2^(n+1) - 2^n - 1
                = 2^n - 1 > 0 для всех n ≥ 1
Значит, последовательность строго возрастающая.

b. Ограниченность:
Последовательность не ограничена сверху, так как 2^n растет быстрее, чем n.

c. Пятый член:
a5 = 2^5 - 5 = 32 - 5 = 27

2) {bn}∞n=2 = 1 / (1-n)

a. Монотонность:
Рассмотрим разность bn+1 - bn = 1/(1-(n+1)) - 1/(1-n) = 1/(-n) - 1/(1-n)
                = (1-n-(-n))/(n(n-1)) = 1/(n(n-1)) > 0 для n > 1
Значит, последовательность строго возрастающая.

b. Ограниченность:
Последовательность не ограничена ни сверху, ни снизу, так как стремится к -∞ при n→2+ и к +∞ при n→+∞.

c. Пятый член:
b5 = 1 / (1-5) = -1/4

3) {cn}∞n=1 = -1^n + √(2n)

a. Монотонность:
Последовательность не является монотонной, так как чередуется между положительными и отрицательными значениями из-за -1^n.

b. Ограниченность:
Последовательность не ограничена сверху, так как √(2n) неограниченно растет.

c. Пятый член:
c5 = -1^5 + √(2*5) = -1 + √10

4) {dn}∞n=1 = (-1)^(2n) + 1/n^2

a. Монотонность:
(-1)^(2n) всегда равно 1, поэтому последовательность эквивалентна 1 + 1/n^2, которая строго убывающая.

b. Ограниченность:
Последовательность ограничена сверху числом 2 и снизу числом 1.

c. Пятый член:
d5 = (-1)^(2*5) + 1/5^2 = 1 + 1/25 = 26/25

```python
# 3

from decimal import Decimal, getcontext

def lim(n):
    getcontext().prec = 8  # устанавливаем точность в 7 знаков после запятой
    
    result = Decimal(1)
    fact = Decimal(1)
    
    for i in range(1, n+1):
        fact *= Decimal(i)
        term = Decimal(1) / Decimal(fact).sqrt()
        result += term
        
        if term < Decimal(1e-7):
            break
        
    return float(result)

print(lim(100))  # вычисляем предел с n=100
```

```python
#4

from decimal import Decimal, getcontext
import math

def calculate_limit_optimized(epsilon):
    getcontext().prec = 8
    
    n = Decimal(1)
    prev_result = Decimal(0)
    
    while True:
        log_numerator = n.ln()
        log_denominator = Decimal(sum(math.log(i) for i in range(1, int(n)+1))) / n
        result = (log_numerator - log_denominator).exp()
        
        if abs(result - prev_result) < epsilon:
            return result
        
        prev_result = result
        n += 1

# Вычисляем предел с точностью 10^-7
epsilon = Decimal('1E-7')
limit = calculate_limit_optimized(epsilon)
print(f"Предел (оптимизированный): {limit}")
```

Как видно из таблицы, формула (A ∨ B) → (B ∨ ¬A) является тавтологией, поскольку она истинна для всех комбинаций истинностных значений A и B.
